EXPLORATORY_DATA_ANALYSIS

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot a histogram of the ages
plt.hist(df['age'], bins=20, color='purple', width=0.6)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
# Create a histogram of student counts based on parent income
plt.hist(df['parentincome'], bins=10, color='c')

plt.xlabel('Parent_Income')
plt.ylabel('Student_Count')

plt.title('Distribution of Student Counts Based on Parent Income')
plt.show()
```

```python
# create a bar chart to compare gender distribution
gender_counts = df['gender'].value_counts()
colors = ['hotpink', 'black']
plt.bar(gender_counts.index, gender_counts.values, color=colors)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# create a bar chart to compare region distribution
region_counts = df['region'].value_counts()
colors = ['hotpink', 'black']
plt.bar(region_counts.index, region_counts.values, color=colors)
plt.title('Region Distribution')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
# create a bar chart to compare internet access distribution
internet_counts = df['internet'].value_counts()
colors = ['black', 'hotpink']
plt.bar(internet_counts.index, internet_counts.values, color=colors)
plt.title('Internet Access Distribution')
plt.xlabel('Internet Access')
plt.ylabel('Count')
plt.show()
```

```python
# heatmap
corr_columns = ['age', 'traveltime', 'studytime', 'performance', 'parentincome', 'freetime', 'attendance', 'G1', 'G2', 'G3']

#  correlation matrix
corr = df[corr_columns].corr()

sns.heatmap(corr, annot=True, cmap='magma')
```
