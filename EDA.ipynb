EXPLORATORY_DATA_ANALYSIS

 There are a large number of possibilities when it comes to analyzing the data using visualization.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

➢ AGE: 

```python
# Plot a histogram of the ages
plt.hist(df['age'], bins=20, color='purple', width=0.6)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

 ➢ AGE: 
 Histograms:
 To visualize the distribution of a continuous variable. The majority of the students in the dataset are in the age range of 16- 17, years old, with a combined frequency of approximately 200. 
 The least represented age group in the dataset is 21-22 years old, with  only 2 students falling in that age range. There is a noticeable drop in frequency for the age group of 19-20 years  old, with only around 25 students falling in that range. This could  indicate that there is a particular reason for this drop in the data, and it  may be worth exploring further.


➢ PARENT_ICOME:

```python
# Create a histogram of student counts based on parent income
plt.hist(df['parentincome'], bins=10, color='c')

plt.xlabel('Parent_Income')
plt.ylabel('Student_Count')

plt.title('Distribution of Student Counts Based on Parent Income')
plt.show()
```

➢ PARENT_ICOME:
The histogram chart shows the distribution of students based  on their parents' income. The x-axis represents the income range, and  the y-axis frequency or count of students falling in that income range.  
 The chart has several distinct bars, with the first bar showing a  frequency of over 150 students - income range between 65,000 and  100,000. The next set of bars represent a continuous range of incomes  between 200,000 - 300,000, frequency above 20 and around 40. The  final set of bars also represent a continuous range of incomes between  400,000 and 500,000, with three bars showing a frequency above 35  and around 40. 

➢ GENDER: 

```python
# create a bar chart to compare gender distribution
gender_counts = df['gender'].value_counts()
colors = ['hotpink', 'black']
plt.bar(gender_counts.index, gender_counts.values, color=colors)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

➢ GENDER: 
Bar charts: To compare categorical variables 
Based on the bar graph, we can infer that there are slightly more females  than males in the dataset. The count of females is 210 that is 55%, while  the count of males is 195 which is 45%. The percentage difference  among male and female students is 10percent.
However, the difference is not significant, and the gender distribution  can be considered relatively balanced. This information can be useful in  designing targeted mentorship and support programs for both genders. 

➢ REGION:

```python
# create a bar chart to compare region distribution
region_counts = df['region'].value_counts()
colors = ['hotpink', 'black']
plt.bar(region_counts.index, region_counts.values, color=colors)
plt.title('Region Distribution')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

➢ REGION:
The bar plot shows that the majority of students in the dataset are from  urban areas (around 77% of the total). Based bar plot, we can see that  the majority of students belong to the urban region (310), while a smaller  number of students belong to the rural region (90). This indicates that there is a higher concentration of students in urban areas compared to  rural areas. This information can be useful for policymakers to target  programs and resources to support students in rural areas.

➢ INTERNET FACILITY:

```python
# create a bar chart to compare internet access distribution
internet_counts = df['internet'].value_counts()
colors = ['black', 'hotpink']
plt.bar(internet_counts.index, internet_counts.values, color=colors)
plt.title('Internet Access Distribution')
plt.xlabel('Internet Access')
plt.ylabel('Count')
plt.show()
```

➢ CORRELATION:

```python
# heatmap
corr_columns = ['age', 'traveltime', 'studytime', 'performance', 'parentincome', 'freetime', 'attendance', 'G1', 'G2', 'G3']

#  correlation matrix
corr = df[corr_columns].corr()

sns.heatmap(corr, annot=True, cmap='magma')
```

➢ CORRELATION:
Heatmaps to visualize the correlation between  multiple variables 
This is a heatmap showing the correlation matrix for various variables in  the dataset age, travel time, study time, performance, parent income, free  time, attendance, G1, G2, and G3.
The color intensity and the numbers inside the squares show the strength  and direction of the correlation between the variables. The variables with  a strong positive correlation will have a deeper shade of red, while  variables with a strong negative correlation will have a deeper shade of  blue. The variables with no correlation or a weak correlation will be  shown in a lighter color or with no color at all. The annotations inside  the squares show the correlation coefficient, which ranges from -1 to +1,  with -1 indicating a strong negative correlation, 0 indicating no  correlation, and +1 indicating a strong positive correlation.
